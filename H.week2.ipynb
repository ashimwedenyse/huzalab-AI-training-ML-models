```python
def digitCount(n):
    if n == 0:
        return 1
    count = 0
    n = abs(n)
    while n > 0:
        n //= 10
        count += 1

    return count
print(digitCount(12323))
print(digitCount(0))
print(digitCount(-111))
```

```python
def gcd(x, y):
    if y == 0:
        return x

    #  apply Euclid's algorithm in recursion function
    return gcd(y, x % y)
print(gcd(270, 250))
print(gcd(48, 18))
print(gcd(17, 5))
print(gcd(10, 5))
```

```python
def hasConsecutiveDigits(n):
    n_str = str(abs(n))
    for i in range(len(n_str) - 1):
        if n_str[i] == n_str[i + 1]:
            return True

    return False
print(hasConsecutiveDigits(123))
print(hasConsecutiveDigits(1223))
print(hasConsecutiveDigits(-1223))
print(hasConsecutiveDigits(12345))
```

```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def sum_of_digits(num):
    return sum(int(digit) for digit in str(num))

def nthAdditivePrime(n):
    count = 0
    num = 2
    while True:
        if is_prime(num) and is_prime(sum_of_digits(num)):
            count += 1
            if count == n:
                return num
        num += 1
print(nthAdditivePrime(1))
print(nthAdditivePrime(2))
print(nthAdditivePrime(3))
print(nthAdditivePrime(4))
```

```python
def mostFrequentDigit(n):
    digit_freq = {i: 0 for i in range(10)}
    for digit in str(abs(n)):
        digit_freq[int(digit)] += 1
    max_freq = max(digit_freq.values())
    for digit in range(10):
        if digit_freq[digit] == max_freq:
            return digit
print(mostFrequentDigit(123456))
print(mostFrequentDigit(-112233))
print(mostFrequentDigit(987654321))
```

```python
def Rotation(x, y):
    x_str = str(x)
    y_str = str(y)
    if len(x_str) != len(y_str):
        return False
    y_str += y_str
    if x_str in y_str:
        return True
    else:
        return False
print(Rotation(3412, 1234))
print(Rotation(1234, 3412))
print(Rotation(1234, 5678))
print(Rotation(1234, 1243))
```

```python
def integral(f, a, b, N):
    """
    Approximates the area under the curve of the function f(x) from x=a to x=b
    using the trapezoidal rule with N trapezoids.

    Parameters:
    f : function
        The function for which the area under the curve is to be approximated.
    a : float
        The lower bound of integration.
    b : float
        The upper bound of integration.
    N : int
        The number of trapezoids to use in the approximation.

    Returns:
    float
        The approximate area under the curve of f(x) from x=a to x=b.
    """
    width = (b - a) / N
    area = 0

    for i in range(N):
        x0 = a + i * width
        x1 = a + (i + 1) * width
        area += (f(x0) + f(x1)) * width / 2

    return area
def f1(x):
    return x

def f2(x):
    return 2 * x + 3

def f3(x):
    return 2 * x ** 2

print(integral(f1, 0, 1, 10))
print(integral(f2, 0, 1, 10))
print(integral(f3, 0, 1, 10))



```

```python
def findZeroWithBisection(f, x0, x1, epsilon):

    if f(x0) * f(x1) >= 0:
        raise ValueError("Initial values must have opposite signs")

    x_mid = (x0 + x1) / 2

    while abs(x1 - x0) > epsilon:

        if f(x_mid) == 0:
            return x_mid


        if f(x0) * f(x_mid) < 0:
            x1 = x_mid
        else:
            x0 = x_mid
        x_mid = (x0 + x1) / 2

    return x_mid

def f(x):
    return x**2 - 4

root = findZeroWithBisection(f, 0, 3, 0.001)
print("Approximate root:", root)
```

```python
def prime_factors(n):
    factors = []
    divisor = 2
    while divisor <= n:
        if n % divisor == 0:
            factors.append(divisor)
            n //= divisor
        else:
            divisor += 1
    return factors

def digit_sum(num):
    return sum(int(digit) for digit in str(num))

def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

def nthSmithNumber(n):
    count = 0
    num = 4
    while True:
        if not is_prime(num):
            prime_factors_sum = sum(digit_sum(pf) for pf in prime_factors(num))
            if prime_factors_sum == digit_sum(num):
                count += 1
                if count == n:
                    return num
        num += 1
n = int(input("Enter the value of n: "))
result = nthSmithNumber(n)
print(f"The {n}th Smith number is: {result}")
```

```python
import random

def roll_dice():
    return random.randint(1, 6)

def playPig():
    print("Welcome to the Pig game!")

    player1_name = input("Enter the name of Player 1: ")
    player2_name = input("Enter the name of Player 2: ")

    player1_score = 0
    player2_score = 0

    current_player = 1

    while True:
        print(f"\n{player1_name}'s score: {player1_score}")
        print(f"{player2_name}'s score: {player2_score}")

        if current_player == 1:
            player_name = player1_name
        else:
            player_name = player2_name

        print(f"\n{player_name}'s turn!")

        roll_or_hold = input("Do you want to roll (r) or hold (h)? ")

        if roll_or_hold.lower() == 'r':
            dice_roll = roll_dice()
            print(f"You rolled: {dice_roll}")
            if dice_roll == 1:
                print("You rolled a 1! You lose all points for this turn.")
                if current_player == 1:
                    player1_score = 0
                else:
                    player2_score = 0
                current_player = 3 - current_player
            else:
                if current_player == 1:
                    player1_score += dice_roll
                else:
                    player2_score += dice_roll
        elif roll_or_hold.lower() == 'h':
            print(f"{player_name} chose to hold.")
            current_player = 3 - current_player

        if player1_score >= 100:
            print(f"{player1_name} wins with a score of {player1_score}!")
            break
        elif player2_score >= 100:
            print(f"{player2_name} wins with a score of {player2_score}!")
            break

playPig()
```
