1.BOOK Class



```python
class Book:
    def __init__(self, title, author, isbn, availability=True):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.__availability = availability

    def get_availability(self):
        return self.__availability

    def set_availability(self, availability):
        self.__availability = availability

    def display_info(self):
        print("Title:", self.title)
        print("Author:", self.author)
        print("ISBN:", self.isbn)
        print("Availability:", "Available" if self.__availability else "Not Available")

if __name__ == "__main__":
    book1 = Book("Python Programming", "John Doe", "978-0-13-267820-9")
    book1.display_info()

    print("\nUpdating availability...")
    book1.set_availability(False)
    book1.display_info()

```

2.Library Class

```python
class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)
        print(f"Book '{book.title}' added to the library.")

    def remove_book(self, book):
        if book in self.books:
            self.books.remove(book)
            print(f"Book '{book.title}' removed from the library.")
        else:
            print("Book not found in the library.")

    def list_all_books(self):
        if self.books:
            print("Books available in the library:")
            for book in self.books:
                print(f"- {book.title} by {book.author}")
        else:
            print("No books available in the library.")

    def search_book(self, keyword):
        found_books = []
        for book in self.books:
            if keyword.lower() in book.title.lower() or keyword.lower() in book.author.lower():
                found_books.append(book)
        if found_books:
            print("Matching books found:")
            for book in found_books:
                print(f"- {book.title} by {book.author}")
        else:
            print("No matching books found in the library.")


class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn

if __name__ == "__main__":
    library = Library()

    book1 = Book("Python Programming", "John Doe", "978-0-13-267820-9")
    book2 = Book("Introduction to Algorithms", "Thomas H. Cormen", "978-0-262-03384-8")
    library.add_book(book1)
    library.add_book(book2)

    library.list_all_books()
    library.search_book("Python")

    library.remove_book(book1)

    library.list_all_books()
```

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.available = True

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def display_books(self):
        for book in self.books:
            print(f"{book.title} by {book.author} - {'Available' if book.available else 'Not Available'}")

    def find_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                return book
        return None

    def check_out_book(self, title):
        book = self.find_book(title)
        if book:
            if book.available:
                book.available = False
                print(f"You have successfully checked out {book.title}")
            else:
                print(f"Sorry, {book.title} is not available for checkout")
        else:
            print(f"Sorry, {title} is not in our library")

    def return_book(self, title):
        book = self.find_book(title)
        if book:
            if not book.available:
                book.available = True
                print(f"Thank you for returning {book.title}")
            else:
                print(f"This book ({book.title}) is already available in the library")
        else:
            print(f"Sorry, {title} is not in our library")

library = Library()
book1 = Book("Harry Potter", "J.K. Rowling")
book2 = Book("The Great Gatsby", "F. Scott Fitzgerald")
library.add_book(book1)
library.add_book(book2)

library.display_books()

library.check_out_book("Harry Potter")
library.check_out_book("The Great Gatsby")
library.check_out_book("The Catcher in the Rye")

library.display_books()

library.return_book("Harry Potter")
library.return_book("The Great Gatsby")

library.display_books()
```

4. User Interaction


```python
class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title, author):
        self.books.append({'title': title, 'author': author, 'available': True})
        print(f"Book '{title}' by {author} has been added to the library.")

    def list_books(self):
        if not self.books:
            print("No books available in the library.")
        else:
            print("Books available in the library:")
            for i, book in enumerate(self.books, 1):
                availability = "Available" if book['available'] else "Borrowed"
                print(f"{i}. '{book['title']}' by {book['author']} [{availability}]")

    def borrow_book(self, title):
        for book in self.books:
            if book['title'].lower() == title.lower() and book['available']:
                book['available'] = False
                print(f"You have borrowed '{book['title']}' by {book['author']}")
                return
        print("Sorry, the book is either not available or doesn't exist in the library.")

    def return_book(self, title):
        for book in self.books:
            if book['title'].lower() == title.lower() and not book['available']:
                book['available'] = True
                print(f"You have returned '{book['title']}' by {book['author']}")
                return
        print("Sorry, the book is either already available or doesn't exist in the library.")

def main():
    library = Library()
    while True:
        print("\nLibrary System Menu:")
        print("1. Add a book")
        print("2. List all books")
        print("3. Borrow a book")
        print("4. Return a book")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == '1':
            title = input("Enter the title of the book: ")
            author = input("Enter the author of the book: ")
            library.add_book(title, author)

        elif choice == '2':
            library.list_books()

        elif choice == '3':
            title = input("Enter the title of the book you want to borrow: ")
            library.borrow_book(title)

        elif choice == '4':
            title = input("Enter the title of the book you want to return: ")
            library.return_book(title)

        elif choice == '5':
            print("Exiting the library system. Thank you!")
            break

        else:
            print("Invalid choice. Please enter a number between 1 and 5.")

if __name__ == "__main__":
    main()
```

5. Inheritance - Extending the System


```python
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages

    def display_info(self):
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
        print(f"Pages: {self.pages}")

class DigitalBook(Book):
    def __init__(self, title, author, pages, file_size, file_format):
        super().__init__(title, author, pages)
        self.file_size = file_size
        self.file_format = file_format

    def display_info(self):
        super().display_info()
        print(f"File Size: {self.file_size}")
        print(f"Format: {self.file_format}")

    def download(self):
        print(f"Downloading {self.title}...")
        print("Download complete.")

book1 = DigitalBook("Python Programming", "John Smith", 300, "2.5 MB", "PDF")
book1.display_info()
book1.download()
```
