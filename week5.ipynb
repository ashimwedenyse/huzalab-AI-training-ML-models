```python
import os

def countFiles(path):
    def count_files_helper(folder_path):
        total_files = 0

        if os.path.isdir(folder_path):
            for item in os.listdir(folder_path):
                item_path = os.path.join(folder_path, item)
                if os.path.isfile(item_path):
                    total_files += 1
                elif os.path.isdir(item_path):
                    total_files += count_files_helper(item_path)

        return total_files

    return count_files_helper(path)



```

```python
import sys

def isPrime(num):
    """ Check if a number is prime """
    if num <= 1:
        return False
    if num <= 3:
        return True
    if num % 2 == 0 or num % 3 == 0:
        return False
    i = 5
    while i * i <= num:
        if num % i == 0 or num % (i + 2) == 0:
            return False
        i += 6
    return True

def findRTP(digits):
    """ Find the smallest RTP with the specified number of digits """
    def backtrack(current, length):
        nonlocal smallest_rtp

        if len(current) == length:
            if isPrime(int(current)):
                smallest_rtp = int(current)
            return

        # Try appending digits from 1 to 9 to form a new number
        for digit in '123456789':
            new_num = current + digit
            if isPrime(int(new_num)):
                backtrack(new_num, length)
                # If the smallest RTP is found, stop further exploration
                if smallest_rtp is not None:
                    return

    smallest_rtp = None
    # Start the backtrack with an empty string and the desired number of digits
    backtrack('', digits)

    return smallest_rtp
print(findRTP(3))
print(findRTP(4))

```

```python
def solvePegSolitaire(board):
    def parse_board(board_str):
        """ Parse the board string into a 2D list representation """
        return [list(row.strip()) for row in board_str.split("\n") if row.strip()]

    def is_valid_move(board, r1, c1, r2, c2):
        """ Check if a move from (r1, c1) to (r2, c2) is valid """
        return 0 <= r1 < len(board) and 0 <= c1 < len(board[0]) and \
               0 <= r2 < len(board) and 0 <= c2 < len(board[0]) and \
               board[r1][c1] == 'O' and board[r2][c2] == '.' and \
               ((r2 == r1 + 2 and c2 == c1) or (r2 == r1 - 2 and c2 == c1) or \
                (r2 == r1 and c2 == c1 + 2) or (r2 == r1 and c2 == c1 - 2))

    def apply_move(board, r1, c1, r2, c2):
        """ Apply a move on the board """
        board[r1][c1] = '.'
        board[(r1 + r2) // 2][(c1 + c2) // 2] = '.'
        board[r2][c2] = 'O'

    def undo_move(board, r1, c1, r2, c2):
        """ Undo a move on the board """
        board[r1][c1] = 'O'
        board[(r1 + r2) // 2][(c1 + c2) // 2] = 'O'
        board[r2][c2] = '.'

    def find_moves(board):
        """ Find all possible moves on the board """
        moves = []
        for r in range(len(board)):
            for c in range(len(board[0])):
                if board[r][c] == 'O':
                    for dr, dc in [(2, 0), (-2, 0), (0, 2), (0, -2)]:
                        nr, nc = r + dr, c + dc
                        if is_valid_move(board, r, c, nr, nc):
                            moves.append((r, c, nr, nc))
        return moves

    def dfs(board, depth, max_depth, path):
        """ Depth-first search with backtracking """
        if depth > max_depth:
            return None

        if sum(row.count('O') for row in board) == 1:
            return path

        moves = find_moves(board)
        for r1, c1, r2, c2 in moves:
            apply_move(board, r1, c1, r2, c2)
            path.append((r1, c1, r2, c2))

            result = dfs(board, depth + 1, max_depth, path)
            if result:
                return result

            path.pop()
            undo_move(board, r1, c1, r2, c2)

    # Parse the board string into a 2D list
    board = parse_board(board)

    # Start with a reasonable maximum depth based on board size
    max_depth = len(board) * len(board[0])

    # Attempt to solve the puzzle using DFS with backtracking
    solution = dfs(board, 0, max_depth, [])

    return solution

# Example usage:
board10 = """\
  ...
  .O.
..OO.O.
.O...O.
..O.O..
  O.O
  ...
"""

result = solvePegSolitaire(board10)
print(result)



```

```python
def oddCount(L):
    """ Return the number of odd integers in the list L """
    if not L:
        return 0  # Base case: empty list has 0 odd integers

    # Count odd numbers in the first half and the second half of the list
    first_half_count = oddCount(L[:len(L)//2])
    second_half_count = oddCount(L[len(L)//2:])

    # Combine counts from both halves
    return first_half_count + second_half_count


print(oddCount([]))
print(oddCount([1, 2, 3, 4, 5]))
print(oddCount([2, 4, 6, 8]))
print(oddCount([1, 1, 1, 1, 1]))
print(oddCount([2, 3, 5, 7, 11]))
```

```python
def oddSum(L):
    """ Return the sum of odd integers in the list L """
    if not L:
        return 0  # Base case: empty list has sum of 0

    first = L[0]
    rest = L[1:]

    # Check if the first element is odd
    if first % 2 != 0:
        return first + oddSum(rest)

    return oddSum(rest)

print(oddSum([]))
print(oddSum([1, 2, 3, 4, 5]))
print(oddSum([2, 4, 6, 8]))
print(oddSum([1, 1, 1, 1, 1]))
print(oddSum([2, 3, 5, 7, 11]))
```

```python
def oddsOnly(L):
    """ Return a new list containing only the odd integers from list L """
    if not L:
        return []

    first = L[0]
    rest = L[1:]

    if first % 2 != 0:
        # If the first element is odd, include it in the result list
        return [first] + oddsOnly(rest)
    else:
        # If the first element is even, exclude it and continue with the rest
        return oddsOnly(rest)
print(oddsOnly([]))
print(oddsOnly([1, 2, 3, 4, 5]))
print(oddsOnly([2, 4, 6, 8]))
print(oddsOnly([1, 1, 1, 1, 1]))
print(oddsOnly([2, 3, 5, 7, 11]))
```

```python
def maxOdd(L):
    """ Return the largest odd integer in list L, or None if no odd integers """
    if not L:
        return None

    first = L[0]
    rest = L[1:]
    max_in_rest = maxOdd(rest)
    if first % 2 != 0:
        if max_in_rest is None or first > max_in_rest:
            return first
        else:
            return max_in_rest
    else:

        return max_in_rest

print(maxOdd([]))
print(maxOdd([1, 2, 3, 4, 5]))
print(maxOdd([2, 4, 6, 8]))
print(maxOdd([1, 1, 1, 1, 1]))
print(maxOdd([2, 3, 5, 7, 11]))
```

```python
def hasConsecutiveDigits(n):
    """ Return True if the integer n contains two consecutive identical digits, False otherwise """
    s = str(n)
    if len(s) < 2:
        return False
    if s[0] == s[1]:
        return True
    else:
        # Recursively check the rest of the string
        return hasConsecutiveDigits(int(s[1:]))

# Test cases
print(hasConsecutiveDigits(123))
print(hasConsecutiveDigits(112233))
print(hasConsecutiveDigits(-112233))
print(hasConsecutiveDigits(9876543210))
print(hasConsecutiveDigits(122345))
```

```python
def alternatingSum(L):
    """ Return the alternating sum of the list L """
    def alternatingSumHelper(lst, subtract_next):
        # Base case: empty list, return 0
        if not lst:
            return 0

        first = lst[0]
        rest = lst[1:]

        if subtract_next:
            current_sum = -first
        else:
            current_sum = first


        return current_sum + alternatingSumHelper(rest, not subtract_next)


```
