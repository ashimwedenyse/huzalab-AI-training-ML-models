```python
import matplotlib.pyplot as plt
import seaborn as sns

#a histogram for maize production
sns.set(style="whitegrid")
plt.figure(figsize=(10, 5))
plt.hist(df['Total Production (Ton)'], 20,  edgecolor='black')
plt.title('Distribution of Maize Production (Ton)')
plt.xlabel('Total Production (Ton)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a boxplot for maize production
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Total Production (Ton)'])
plt.title('Boxplot of Maize Production (Ton)')
plt.xlabel('Total Production (Ton)')
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Load the shapefile of Rwanda's districts
rwanda_districts = gpd.read_file('rwa_district')

# Merge the GeoDataFrame with the maize production DataFrame
merged_df = rwanda_districts.set_index('District').join(df.set_index('District'))

# Plot the map
fig, ax = plt.subplots(1, 1, figsize=(10, 10))

# Plot the districts with maize production data
merged_df.plot(column='Total Production (Ton)', ax=ax, legend=True,
               legend_kwds={'label': "Maize Production (Ton)",
                            'orientation': "horizontal"}, cmap='Oranges')



for idx, row in merged_df.iterrows():
   
    plt.annotate(row.name, xy=(row.geometry.centroid.x, row.geometry.centroid.y),
                 horizontalalignment='center', fontsize=8, color='black')


# Add title and labels
plt.title('Maize Production by District in Rwanda', fontsize=15)
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Show the plot
plt.show()
```
