```python
import math

def distance(x1, y1, x2, y2):
    return math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)

x1 = 1
y1 = 2
x2 = 4
y2 = 6
print(distance(x1, y1, x2, y2))
```

```python
import math

def circlesIntersect(x1, y1, r1, x2, y2, r2):
    distance_between_centers = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
    sum_of_radii = r1 + r2
    return distance_between_centers <= sum_of_radii

x1 = 0
y1 = 0
r1 = 3
x2 = 5
y2 = 0
r2 = 4
print(circlesIntersect(x1, y1, r1, x2, y2, r2))
```

```python
def getInRange(x, bound1, bound2):
    lower_bound = min(bound1, bound2)
    upper_bound = max(bound1, bound2)

    if x < lower_bound:
        return lower_bound
    elif x > upper_bound:
        return upper_bound
    else:
        return x
print(getInRange(1, 3, 5))
print(getInRange(4, 3, 5))
print(getInRange(6, 3, 5))
print(getInRange(6, 5, 3))
```

```python
import math

def eggCartons(eggs):
    return math.ceil(eggs / 12)

print(eggCartons(20))
print(eggCartons(35))
```

```python
import math

def pascalsTriangleValue(row, col):
    if row < 0 or col < 0 or col > row:
        return None

    return math.factorial(row) // (math.factorial(col) * math.factorial(row - col))

print(pascalsTriangleValue(0, 0))
print(pascalsTriangleValue(3, 2))
print(pascalsTriangleValue(5, 2))
```

```python
def getKthDigit(n, k):
    n_str = str(abs(n))
    if k < 0 or k >= len(n_str):
        return None
    digit = int(n_str[-(k+1)])
    if n < 0:
        digit *= -1

    return digit

# Example usage:
print(getKthDigit(12345, 2))
print(getKthDigit(-9876, 1))
print(getKthDigit(123, 0))
print(getKthDigit(789, 3))
```

```python
def setKthDigit(n, k, d):
    digits = [int(i) for i in str(abs(n))]
    if k < 0 or k >= len(digits):
        return None
    digits[-(k+1)] = d
    result = int(''.join(map(str, digits)))
    if n < 0:
        result *= -1

    return result
print(setKthDigit(12345, 2, 9))
print(setKthDigit(-9876, 1, 3))
print(setKthDigit(123, 0, 8))
print(setKthDigit(789, 3, 0))
```

```python
# partB
def nearestOdd(n):
    n_int = int(n)
    if n_int % 2 != 0:
        return n_int
    lower_odd = n_int - 1
    upper_odd = n_int + 1

    if abs(lower_odd - n) < abs(upper_odd - n):
        return lower_odd
    else:
        return upper_odd

# Example usage:
print(nearestOdd(13.0))
print(nearestOdd(14))
print(nearestOdd(15))
print(nearestOdd(16.5))
print(nearestOdd(17.5))
```

```python
def blend_colors(color1, color2, ratio):
    """
    Blend two RGB colors based on the specified blend ratio.

    Arguments:
    color1: Tuple representing the first RGB color (r1, g1, b1).
    color2: Tuple representing the second RGB color (r2, g2, b2).
    ratio: Float representing the blend ratio (between 0 and 1).

    Returns:
    Tuple representing the blended RGB color.
    """
    r1, g1, b1 = color1
    r2, g2, b2 = color2

    blended_r = int(r1 * (1 - ratio) + r2 * ratio)
    blended_g = int(g1 * (1 - ratio) + g2 * ratio)
    blended_b = int(b1 * (1 - ratio) + b2 * ratio)

    return (blended_r, blended_g, blended_b)

color1 = (255, 0, 0)
color2 = (0, 0, 255)
blend_ratio = 0.5
blended_color = blend_colors(color1, color2, blend_ratio)
print("Blended color:", blended_color)
```
