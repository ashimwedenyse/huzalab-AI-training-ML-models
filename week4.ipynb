week8. extrapractice question1 finding most common name

```python
def mostCommonName(L):
    if not L:
        return None

    name_count = {}
    max_count = 0

    # Count occurrences of each name
    for name in L:
        name_count[name] = name_count.get(name, 0) + 1
        max_count = max(max_count, name_count[name])

    # Find names with the maximum count
    most_common_names = {name for name, count in name_count.items() if count == max_count}

    if len(most_common_names) == 1:
        return most_common_names.pop()
    else:
        return most_common_names

# Test function
def testMostCommonName():
    print("Testing mostCommonName()...", end="")
    assert(mostCommonName(["Jane", "Aaron", "Cindy", "Aaron"]) == "Aaron")
    assert(mostCommonName(["Jane", "Aaron", "Jane", "Cindy", "Aaron"]) == {"Aaron", "Jane"})
    assert(mostCommonName(["Cindy"]) == "Cindy")
    assert(mostCommonName(["Jane", "Aaron", "Cindy"]) == {"Aaron", "Cindy", "Jane"})
    assert(mostCommonName([]) == None)
    print("Passed!")

testMostCommonName()
```

Most popular friend question5

```python
def mostPopularFriend(d):
    friend_count = {}  # Dictionary to store the counts of each friend

    # Count occurrences of each friend
    for friends_set in d.values():
        for friend in friends_set:
            friend_count[friend] = friend_count.get(friend, 0) + 1

    # Find the friend with the highest count
    most_popular_friend = max(friend_count, key=friend_count.get)

    return most_popular_friend

# Test the function
d = {}
d["fred"] = set(["wilma", "betty", "barney"])
d["wilma"] = set(["fred", "betty", "dino"])

print(mostPopularFriend(d))  # Output: "betty"
```

Triplets question6

```python
def findTriplets(L):
    triplets = set()

    # Convert the list to a set for faster lookups
    L_set = set(L)

    # Iterate through each pair of elements in the list
    for i in range(len(L) - 1):
        for j in range(i + 1, len(L)):
            # Calculate the complement needed to make the sum zero
            complement = -(L[i] + L[j])

            # Check if the complement exists in the set and if it's not one of the current elements
            if complement in L_set and complement != L[i] and complement != L[j]:
                # Add the triplet to the set (ordered to avoid duplicates)
                triplet = tuple(sorted([L[i], L[j], complement]))
                triplets.add(triplet)

    return triplets

# Example usage:
L = [-1, 0, -3, 2, 1]
print(findTriplets(L))  # Output: {(-1, 0, 1), (-3, 1, 2)}
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.friends = []

    def getName(self):
        return self.name

    def getAge(self):
        return self.age

    def getFriends(self):
        return self.friends

    def getFriendsNames(self):
        return sorted([friend.getName() for friend in self.friends])

    def addFriend(self, person):
        if person not in self.friends:
            self.friends.append(person)
            person.addFriend(self)

    def addFriends(self, people):
        for person in people:
            self.addFriend(person)


def testPersonClass():
    print('Testing Person Class...', end='')
    fred = Person('fred', 32)
    assert(isinstance(fred, Person))
    assert(fred.getName() == 'fred')
    assert(fred.getAge() == 32)
    assert(fred.getFriends() == [])
    assert(fred.getFriendsNames() == [])

    wilma = Person('wilma', 35)
    assert(wilma.getName() == 'wilma')
    assert(wilma.getAge() == 35)
    assert(wilma.getFriends() == [])

    wilma.addFriend(fred)
    assert(wilma.getFriends() == [fred])
    assert(wilma.getFriendsNames() == ['fred'])
    assert(fred.getFriends() == [wilma])  # friends are mutual!
    assert(fred.getFriendsNames() == ['wilma'])

    wilma.addFriend(fred)
    assert(wilma.getFriends() == [fred])  # don't add twice!

    betty = Person('betty', 29)
    fred.addFriend(betty)
    assert(fred.getFriendsNames() == ['betty', 'wilma'])

    pebbles = Person('pebbles', 4)
    betty.addFriend(pebbles)
    assert(betty.getFriendsNames() == ['fred', 'pebbles'])

    barney = Person('barney', 28)
    barney.addFriend(pebbles)
    barney.addFriend(betty)
    barney.addFriends(fred.getFriends())  # add ALL of Fred's friends as Barney's friends
    assert(barney.getFriends() == [pebbles, betty, wilma])
    assert(barney.getFriendsNames() == ['betty', 'pebbles', 'wilma'])
    fred.addFriend(wilma)
    fred.addFriend(barney)
    assert(fred.getFriends() == [wilma, betty, barney])
    assert(fred.getFriendsNames() == ['barney', 'betty', 'wilma'])  # sorted!
    assert(barney.getFriends() == [pebbles, betty, wilma, fred])
    assert(barney.getFriendsNames() == ['betty', 'fred', 'pebbles', 'wilma'])
    print('Passed!')


# Run test cases
testPersonClass()
```

get pair sum question 3

```python
def getPairSum(L, target):
    complements = {}  # Dictionary to store complements of elements
    for num in L:
        complement = target - num
        if complement in complements:
            return (complements[complement], num)  # Found a pair
        complements[num] = num  # Add current number to complements
    return None  # No pair found

# Test cases
print(getPairSum([1], 1))  # None
print(getPairSum([5, 2], 7))  # (2, 5) or (5, 2)
print(getPairSum([10, -1, 1, -8, 3, 1], 2))  # (-1, 3) or (3, -1) or (1, 1) or (10, -8) or (-8, 10)
print(getPairSum([10, -1, 1, -8, 3, 1], 10))  # None
```

PythagoreanTriple question4



```python
def containsPythagoreanTriple(L):
    squares = set()
    for num in L:
        squares.add(num ** 2)

    for i in range(len(L)):
        for j in range(i + 1, len(L)):
            sum_of_squares = L[i] ** 2 + L[j] ** 2
            if sum_of_squares in squares:
                return True
    return False

# Test cases
print(containsPythagoreanTriple([1, 3, 6, 2, 5, 1, 4]))  # True
print(containsPythagoreanTriple([1, 3, 6, 2, 1, 4]))     # False
```

```python
def movieAwards(oscarResults):
    awards_count = {}  # Dictionary to store the count of awards for each movie

    for category, movie in oscarResults:
        if movie in awards_count:
            awards_count[movie] += 1
        else:
            awards_count[movie] = 1

    return awards_count

# Test case
oscarResults = {
    ("Best Picture", "Green Book"),
    ("Best Actor", "Bohemian Rhapsody"),
    ("Best Actress", "The Favourite"),
    ("Film Editing", "Bohemian Rhapsody"),
    ("Best Original Score", "Black Panther"),
    ("Costume Design", "Black Panther"),
    ("Sound Editing", "Bohemian Rhapsody"),
    ("Best Director", "Roma")
}

print(movieAwards(oscarResults))
```

friend of friends question6

```python
def friendsOfFriends(d):
    friends_of_friends = {}  # Dictionary to store friends-of-friends

    for person, friends in d.items():
        fofs = set()  # Set to store friends-of-friends for the current person
        for friend in friends:
            fofs.update(d.get(friend, set()))  # Add friends of the current friend
        fofs -= friends  # Remove direct friends from friends-of-friends
        fofs.discard(person)  # Remove self from friends-of-friends
        friends_of_friends[person] = fofs

    return friends_of_friends

# Test case
d = {
    "jon": set(["arya", "tyrion"]),
    "tyrion": set(["jon", "jaime", "pod"]),
    "arya": set(["jon"]),
    "jaime": set(["tyrion", "brienne"]),
    "brienne": set(["jaime", "pod"]),
    "pod": set(["tyrion", "brienne", "jaime"]),
    "ramsay": set()
}

print(friendsOfFriends(d))
```

```python
import pygame
import random

# Define colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
GREEN = (0, 255, 0)
RED = (255, 0, 0)
BLUE = (0, 0, 255)

# Initialize pygame
pygame.init()

# Set the screen dimensions
SCREEN_WIDTH = 800
SCREEN_HEIGHT = 600
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("Sorting Algorithm Visualization")

# Set the clock
clock = pygame.time.Clock()

# Helper function to generate a shuffled list
def generate_shuffled_list():
    nums = list(range(21))  # Numbers from 0 to 20
    random.shuffle(nums)
    return nums

# Helper function to draw the list
def draw_list(nums, highlight=None):
    screen.fill(WHITE)
    block_width = SCREEN_WIDTH // len(nums)
    for i, num in enumerate(nums):
        rect = pygame.Rect(i * block_width, SCREEN_HEIGHT - num * 20, block_width, num * 20)
        color = GREEN if highlight and i in highlight else BLACK
        pygame.draw.rect(screen, color, rect)
    pygame.display.flip()

# Bubble sort algorithm
def bubble_sort(nums):
    n = len(nums)
    for i in range(n):
        for j in range(0, n - i - 1):
            yield nums, [j, j + 1]  # Highlight comparison
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]  # Swap
                yield nums, [j, j + 1]  # Highlight swapped elements

# Main function to run the animation
def bonus_run_sorting_animation():
    nums = generate_shuffled_list()  # Generate shuffled list
    sorting_algorithm = bubble_sort(nums)  # Choose sorting algorithm
    highlight = None  # Initialize highlight
    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_RIGHT:
                    try:
                        nums, highlight = next(sorting_algorithm)
                    except StopIteration:
                        highlight = None
                elif event.key == pygame.K_r:
                    nums = generate_shuffled_list()
                    sorting_algorithm = bubble_sort(nums)
                    highlight = None
                draw_list(nums, highlight)
        clock.tick(10)  # Set the FPS
    pygame.quit()

# Run the animation
bonus_run_sorting_animation()
```
